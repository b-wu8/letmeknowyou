To better understand the previous work done on these datasets and try to improve upon them, we decided to learn by implementing. 

- For FER2013 dataset, we implemented three of the best performing models from scratch and fine-tuned them to the best of our abilities. 

  - CNN with Mega tunning.
    - As introduced in lecture, CNN Model tuning was the first idea we had to improve our model compared with the baseline CNN model
    - Due to the low accuracy on both train set and validation set, we realized that complex model structure might be needed for improve the model. Thus we had one more convolution layer with the filter size tuned, and four batch norm layers to let the model be more sensitive to errors in the process of Back-Propagation. Besides, adding a flatten layer was also considered to be transition from convolution layers to fully connected layers
    - After tuning all above, from some visualizations of our baseline model, the model overfitted the training set while the test set loss stops decreasing; therefore we added four layers of dropout layers and a regularizer to avoid the situation.
    - For better training speed and the device type issue, we used tensorflow to implement this model
  - ResNet with modified structure
    - ResNet is a convolutional neural network that can be utilized as a state of the art image classification model. Our ResNet model is implemented from scratch using PyTorch. We choose ResNet for performance comparison because it is a classic CNN structure that gives decent performance and controls overfitting.
    - ResNet is built with skip connections with nonlinearities (ReLU) and batch normalization in between. Building skip connections in between layers avoids the problem of vanishing gradient and makes it easier to optimize neural networks, where the gating mechanisms facilitate information flow across layers. It also mitigates the Degradation (accuracy saturation) problem, where adding more layers to a deep model leads to higher training error.
    - We choose ResNet because we want to see how these special skip connections can be effective in improving the performance as well as controlling overfitting. 
    - Deep layered ResNet is difficult to implement from scratch, so we decided to implement a lighter layered ResNet that is not significantly down in performance.

  - VGG16 replica with modified linear layers
    - Another algorithm that we have tried is VGG16. The main reason why we choose this model is that it is still considered to be one of the best classic models, and we are not strangers to this model since TA Sun Shuo have us walked through this model in his discussion section. 

    - This model is relatively easy to implement from scratch using Pytorch since the structure is very straight forward. However this is a pretty large network and it has about 138 million (approx) parameters, thus it results in a long time of training. 

    - For this model, we keep the convolutional block structures intact, and only tune the parameters and structures in the linear layers. In linear layers, we largely reduce the output channels, and in the last linear layer we changed the output class from 1000 to 7 (7 emotions in our case). However we realized that the model suffers a lot from overfitting. We evaluate the model by accuracy and loss, and the train loss keeps reducing while the validate loss stays the same. So We add drop out stages in our linear structures, but suprisingly adding them reduces the overall performance significantly.

- For Adience dataset age prediction, we implement one baseline model alexnet, and a better model called ConvNext.

    - AlexNet model shares the same structure as our baseline model, but it performs poorly in doing age predictions. Given the results of the AlexNet, obviously we need a more powerful model that can provide us with better results.

    - We decided to implement one of the latest model ConvNext, which is published in January 2022, and we use this model becuase it performs very well in our problem. Instead of embracing the trend of introducing the Vision transformers, previously introduced to solve problems in NLP, that took over the state-of-the-art CNN in solving image classification tasks, ConvNext still builds a pure convolutional neural network whose performance match state-of-the-art-models like the Vision transformer. As described by the authors of ConvNext, they take a standard neural network, a ResNet-50, as a starting point and gradually "modernize" it toward the design of a vision Transformer. In some situation it even surpasses the Transformer in terms of performance.

    - We used a **pretrained** ConvNext model since this model is very complicated to implement from scratch, and the techniques used in constructing this model are beyond the scope of this course, and also as said from professor Dredze "there are lots of other people have built excellent toolkits for machine learning,... and there are so many software packages out there already. There is no sense in building something if someone else has already built it." We believe using a pretrained model will save us a ton of time.

    - Because it is a pretrained model, we do not have many parameters that are tunable. We only changed the structure of the last linear layer and modify the output channel to be 8 age groups.

- For Adience dataset gender prediction, we implement one baseline model in CNN and LeviHassner CNN.

  - This simple CNN baseline model comes directly from our previous baseline model. It performs surprisingly well on the Adience gender classification task. Although the simple CNN overfits greatly in training, it performs fairly well in validation. What we found most surprising is that our baseline model outperforms our devised LeviHassner CNN. Our guess is that our simple CNN keeps the features of the image and does not do pooling or batch norm to disrupt coefficient. Please find more about the results in the Results section below.
  - The LeviHassner CNN is a model proposed by Levi and Hassner in CVPR 2015 in the paper *Age and Gender Classification using Convolutional Neural Networks*. This CNN has local response norm layer between convolutional layers that retain features. In the mean time, it is very easy to implement. We are surprised by how well our baseline model performs in this case.


We evaluate our models based on cross entropy loss and accuracy. The accuracy is the best indicator to tell the model performance and the loss is the best indicator to tell how well and how fast the model has been learning.

- What about these results surprised you? Why?

  - In the Adience gender classification task, our baseline model outperforms our LeviHassner CNN. The major difference in terms of network structure is the connection between CNN layers. The LeviHassner CNN uses local response norm to normalize the data. Our baseline CNN model does not normalize. The LocalResponseNorm() layer most likely destroyed the means and variance of the data and thus affected model performance. This probably means sometimes simpler structure can just perform as well and sometimes more complex structure can lose information while passing between layers. 

  - Our best model for FER2013 is the CNN with hyperparameter tunning. We were surprised by how much performance the baseline model can put out just by hyperparameter tunning. We are suprised by the potential of simple CNN models for easy improvement and easy implementation. Although our implementation of the "better" model could not beat our "boosted" baseline model, we learned a lot in the process of analyzing the model mathematically. 

- Did your models over- or under-fit? How can you tell? What did you do to address these issues?
    - Most of our sentiment analysis models, to some extent, suffer from overfitting, with highest tranning accuracy reaching almost 90% while the validation accuracy barely around 50-60%.
    - Our Adience models are better in dealing with overfitting, and both of our age prediction model and gender prediction model successfully find a balance between bias and variance tradeoff.
    - For overfitting models, we add regularizations to our model, such as drop out stages.

- What does the evaluation of your trained models tell you about your data? How do you expect these models might behave differently on different data?

So we think our models might behave better in high resolution facial images data.
